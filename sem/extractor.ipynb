## USER Record Test for edit and save

```python
with open('sem1.json') as d:
    dictData = json.load(d)
    #print(dictData)
    #print(type(dictData))
sem1df = pd.DataFrame(dictData)
sem1df.drop('Column1',axis=1,inplace=True)
sem1df.index+=1
sem1df
```

## Add Data

```python
def add_data(semdf):
    for i, row in semdf.iterrows():
        grade = str(input(f"Enter Your Grade for {row['Course Title']}: "))
        semdf.at[i,'Your Grade'] = grade
    return semdf
xx = sem1df.iloc[1]

if xx['Your Grade']=='-':
    sem1df = add_data(sem1df)
    sem1df
else:
    user_sem1 = sem1df.head(10)
user_sem1
```

## Edit Data

```python
def edit_data(semdf,course_title,grade):
    i = semdf.index[semdf['Course Title']==course_title].tolist()
    semdf.at[i[0],'Your Grade'] = grade
    return semdf

course_title,grade = str(input("Course Title and new grade to edit :")).split(',')
sem1df = edit_data(sem1df,course_title,grade)
sem1df
```
